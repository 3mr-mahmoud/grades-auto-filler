```python
import pandas as pd

dir = '../../Dataset/Bubble Sheet/'

imgs_path = [dir + '1/1.jpg', dir + '1/2.jpg', dir + '2/1.jpg', dir + '3/1.jpg', dir + '4/1.jpg']
answers_path = [dir + '1/answers.txt', dir + '1/answers.txt', dir + '2/answers.txt', dir + '3/answers.txt', dir + '4/answers.txt']
expecteds_path = [dir + '1/expected1.txt', dir + '1/expected2.txt', dir + '2/expected1.txt', dir + '3/expected1.txt', dir + '4/expected1.txt']

# List to store accuracies
accuracies = []

# Process each image
for img_path, ans_path, exp_path in zip(imgs_path, answers_path, expecteds_path):
    print(f'Processing image: {img_path}')
    try:
        # Open image and answer files
        with open(img_path, 'rb') as image_file, open(ans_path, 'rb') as answers_file:
            df = process_bubble_sheet_gui(image_file, answers_file)
            # print("Detected Results:")
            # print(df)

            if isinstance(df, pd.DataFrame):  # Ensure df is a DataFrame
                # Exclude the 'ID' column for comparison
                result_columns = [col for col in df.columns if col != 'ID']

                # Load expected results from the single-column format
                with open(exp_path, 'r') as f:
                    expected_data = f.readlines()
                # First line is the ID, the rest are the answers
                expected_id = int(expected_data[0].strip())
                expected_answers = [int(answer.strip()) for answer in expected_data[1:]]

                # Create a DataFrame from expected answers for easier comparison
                expected_df = pd.DataFrame([expected_answers], columns=result_columns)
                expected_df.insert(0, 'ID', expected_id)  # Add the ID column
                # print("Expected Results:")
                # print(expected_df)

                # Compare detected results with expected results
                correct_answers = 0
                total_questions = len(result_columns)

                for col in result_columns:
                    correct_answers += int(df[col].iloc[0] == expected_df[col].iloc[0])

                # Calculate accuracy for this image
                if total_questions > 0:
                    accuracy = correct_answers / total_questions * 100
                    accuracies.append((img_path, accuracy))
                    print(f"Accuracy for {img_path}: {accuracy:.2f}%")
                else:
                    print("No questions to process.")
                    accuracies.append((img_path, 0))
            else:
                print(f"Error processing {img_path}: {df}")
                accuracies.append((img_path, 0))
    except Exception as e:
        print(f"Error processing {img_path}: {e}")
        accuracies.append((img_path, 0))

# Print all accuracies
print("\nAccuracies for all images:")
for img, acc in accuracies:
    print(f"{img}: {acc:.2f}%")
```
